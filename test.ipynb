```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load your data (replace with your actual file path)
df = pd.read_csv('sign_mnist_train.csv')

# Function to visualize a single image from the DataFrame
def visualize_image(df, row_index):
    # Extract pixel values (from 'pixel0' to 'pixel783')
    pixel_cols = [col for col in df.columns if col.startswith('pixel')]
    pixels = df.loc[row_index, pixel_cols].values
    
    # Reshape to 28x28
    image = pixels.reshape(28, 28)
    
    # Display the image
    plt.imshow(image, cmap='gray')  # 'gray' for grayscale
    plt.title(f"Label: {df.loc[row_index, 'label']}")  # Show the label
    plt.show()

# Visualize the first image
visualize_image(df, 0)

# Visualize the 10th image
visualize_image(df, 9)

# Visualize multiple images in a grid
def visualize_images_grid(df, num_images=25):
    fig, axes = plt.subplots(5, 5, figsize=(10, 10))  # Adjust figsize as needed
    axes = axes.flatten()

    for i in range(num_images):
        pixels = df.iloc[i, 1:].values  # Skip the 'label' column
        image = pixels.reshape(28, 28)
        axes[i].imshow(image, cmap='gray')
        axes[i].set_title(df.iloc[i, 0], fontsize=8)  # Set title to the label
        axes[i].axis('off')  # Turn off axis labels

    plt.tight_layout()
    plt.show()

# Visualize a grid of images
visualize_images_grid(df, num_images=25)
```
